## rcParams

rcParams are settings for matplotlib. They're how stylesheets work (the thing when you use `plt.style.use`.

You can find examples of the different styles [here](https://matplotlib.org/2.1.1/gallery/style_sheets/style_sheets_reference.html) and [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

> **A big long list is:** bmh, classic, dark_background, fivethirtyeight, ggplot, grayscale, seaborn-bright, seaborn-colorblind, seaborn-dark, seaborn-dark-palette, seaborn-darkgrid, seaborn-deep, seaborn-muted, seaborn-notebook, seaborn-paper, seaborn-pastel, seaborn-poster, seaborn-talk, seaborn-ticks, seaborn-white, seaborn-whitegrid 

If you think one of them is cool, you can find its rcparams styles by checking out its mplstyle file in [in this directory](https://github.com/matplotlib/matplotlib/tree/master/lib/matplotlib/mpl-data/stylelib).

# Grouped bar graph (2 colors)

```python
df = pd.read_csv("gender.csv")
df.head(3)
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
ax = df.plot(x='state', kind='barh', figsize=(13, 7))

# CUSTOMIZATION CODE GOES UNDER HERE

# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0.0,1.0), height=0.12, width=0.02, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

ax.set_title("Gender in American states", pad=40)
ax.title.set_fontsize(22)
ax.grid(True, axis='x')

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'axes.facecolor': '#CDDEE7',
    'figure.facecolor': '#CDDEE7',
    'axes.edgecolor': '#CDDEE7',
    'axes.prop_cycle': cycler('color', ['#00A4DC', '#00526D']),
    'font.sans-serif': 'Adobe Fan Heiti Std',
    'grid.linestyle': '-',
    'grid.linewidth': 2.0,
    'grid.color': '#E2ECF3', 
    'font.size': 15,
    'legend.fancybox': True,
    'xtick.major.size': 7,
    'xtick.minor.size': 7,
    'ytick.major.size': 7,
    'ytick.minor.size': 7,
    'legend.frameon' : False, 
    'legend.handlelength': 1.7,      # the length of the legend lines
    'legend.handleheight': 0.5,
    'axes.labelpad': 8.0,
    'axes.xmargin': .09,
    'axes.ymargin': .09,
    'axes.axisbelow': True
})



# then we save it
#plt.savefig("graph-bar-grouped.pdf")
plt.show()
```

# Line graph

```python
df = pd.read_csv("visas-by-continent-long.csv")
df = df.groupby(["year", "continent"]).visas.sum().unstack().reset_index()
df.drop(columns='Unknown', inplace=True)
df['year'] = pd.to_datetime(df.year, format="%Y")
df.set_index('year', inplace=True)
percent_df = df.div(df.sum(axis=1), axis='rows') * 100

percent_df.head(2)
```

```python
ax = percent_df.plot(kind='line', figsize=(10, 5))

# CUSTOMIZATION CODE GOES UNDER HERE


# Get the figure even though you didn't do the subplots thing
fig = plt.gcf()
# Create a rectangle anchored at 40% of the way across
# and 50% of the way down. height/width grow UP and RIGHT.
rect = plt.Rectangle((0.0,0.9), height=0.12, width=0.02, transform=fig.transFigure, figure=fig, color='red')
# Add it to the figure
fig.patches.append(rect)

ax.set_title("Number of visas by continent")
ax.title.set_fontsize(18)
ax.spines['bottom'].set_visible(True)

plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'axes.facecolor': '#ffffff',
    'figure.facecolor': '#ffffff',
    'axes.edgecolor': '#ffffff',
    'axes.prop_cycle': cycler('color', ['#FB794F', '#E6C774', '#7FCAC7', '#772210', '#79C4DE', '#00485D']),
    'font.sans-serif': 'Adobe Fan Heiti Std',
    'axes.grid': True,
    'grid.linestyle': '-',
    'grid.linewidth': 1.5,
    'grid.color': '#E4E4E5', 
    'font.size': 9,
    'legend.fancybox': True,
    'xtick.major.size': 7,
    'xtick.minor.size': 7,
    'ytick.major.size': 7,
    'ytick.minor.size': 7,
    'legend.frameon' : False, 
    'legend.handlelength': 1.7,      # the length of the legend lines
    'legend.handleheight': 0.5,
    'axes.labelpad': 8.0,
    'axes.xmargin': .09,
    'axes.ymargin': .09,
    'axes.titlepad': 20,
    'legend.borderaxespad': -8,
    'legend.handleheight': 0.2
})


# then we save it
#plt.savefig("graph-lines.pdf")
plt.show()
```
